### GoldHand class
#### Get historical data and prepar for further analytics, also visualize it, with local minimums and maximums
#### It works with stock (for example: TSLA, AAPL) crypto (for example: BTC-USD, ETH-USD) and  ETFs (for example: SPY, VOO)

```python
t = GoldHand('AAPL')
t.df.tail(1).T
```

```python
# display the local min and maximum values
t.df.loc[t.df['local_text']!=''].head(6)
```

```python
# with basic settings
ticker= 'BTC-USD'
t = GoldHand(ticker)
#p = t.plotly_last_year(plot_title = tw.get_plotly_title(ticker))
p = t.plotly_last_year(plot_title = tw.get_plotly_title(ticker), plot_height=800, ndays=1000, ad_local_min_max = True )
p.show()
```

```python
p = t.plotly_last_year(plot_title = tw.get_plotly_title(ticker), plot_height=900, ndays=500, ad_local_min_max=False)
p.show()
```

```python
# with base settings
ticker= 'TSLA'
t = GoldHand(ticker)
#p = t.plot_goldhand_line(tw.get_plotly_title(ticker))
p = t.plot_goldhand_line(tw.get_plotly_title(ticker),plot_height=900, ndays=500, ad_local_min_max=True)
p.show()
```
